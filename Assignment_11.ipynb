```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_anime(anime_title, combined_features_df, df):
    """
    Recommends anime based on cosine similarity of features.

    Args:
        anime_title: The title of the anime to find recommendations for.
        combined_features_df: DataFrame containing the combined numerical and
                              binarized features used for similarity calculation.
        df: The original anime DataFrame containing anime titles.

    Returns:
        A list of recommended anime titles.
    """
    # Check if the input anime title exists in the DataFrame
    if anime_title not in df['name'].values:
        return f"Anime '{anime_title}' not found in the dataset."

    # Get the index of the input anime
    anime_index = df[df['name'] == anime_title].index[0]

    # Calculate cosine similarity
    cosine_sim = cosine_similarity(combined_features_df)

    # Get similarity scores for the input anime
    similarity_scores = list(enumerate(cosine_sim[anime_index]))

    # Sort by similarity scores in descending order
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get top recommendations (excluding the input anime itself)
    recommendations = []
    for i in similarity_scores[1:6]:  # Top 5 recommendations (excluding the anime itself)
        index = i[0]
        recommendations.append(df['name'][index])

    return recommendations
```

**Interview Questions:**

1. Can you explain the difference between user-based and item-based collaborative filtering?

User-Based Collaborative Filtering:

This approach identifies users who are similar to the target user based on their preferences or behavior (e.g., ratings or purchases).
It assumes that if two users are similar, the items liked or rated highly by one user will likely interest the other.
Example: If User A and User B have similar preferences, and User B likes an item that User A hasn't interacted with, that item might be recommended to User A.

Item-Based Collaborative Filtering:

This approach focuses on the similarity between items rather than users. It identifies items that are similar based on how users have interacted with them.
It assumes that if a user likes one item, they are likely to enjoy similar items
Example: If many users who bought Item X also bought Item Y, then Item Y might be recommended to a user who liked Item X.

Difference:

User-based focuses on user similarity to predict preferences, while item-based focuses on item similarity.
Item-based filtering is often more scalable because user preferences tend to change frequently, whereas item relationships remain relatively stable over time.

2. What is collaborative filtering, and how does it work?
   
Collaborative Filtering:

A recommendation system technique that predicts a user's preferences by relying on the past behavior and preferences of many users.
It works by finding patterns in user-item interactions (e.g., ratings, clicks, purchases) to recommend items without requiring explicit domain knowledge about the items.
How It Works:

Data Collection: Gather user-item interaction data, such as ratings, purchase history, or clicks.
Similarity Calculation: Determine similarities between users (user-based) or items (item-based).
Prediction: Use these similarities to predict a user's interest in an item they haven't interacted with yet.
Recommendation: Recommend the top-ranked items to the user.
